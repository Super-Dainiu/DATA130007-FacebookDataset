```python
df = pd.read_csv('./csv/centrality analysis.csv')
```

```python
label = 'n.degree'
ax = plt.gca()
degree=[i[0] for i in pd.DataFrame(df[label]).value_counts().index]
count=np.array(pd.DataFrame(df[label]).value_counts().values)
ax.scatter(degree, count, c=count)
plt.xlabel("Degree")
plt.ylabel("Count")
```
